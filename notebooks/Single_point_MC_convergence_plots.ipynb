# Single point convergence for conformer variable (RMSD), showing how many Monte Carlo iterations are sufficient

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Example single point convergence for molid29 (the PE 16:1 16:1, lipid/lipid like molecule). The "single point" is doing x iterations of the sample size 500 for increasing x

```python
plt.figure(figsize=(10,8))

df = pd.read_csv('../data/SA-rmsd-geoms/molid29+De_single_point_MCconverg_S500_iter100000.csv')


x = df['Iteration']
y = df['Average']
e = df['Standard Dev']

plt.scatter(x, y)
plt.errorbar(x, y, yerr=e, ecolor='#a6aebc')

plt.xscale('log')
plt.title('PE 16:1/16:1 RMSD Single Point Convergence, S=500', size=20)
plt.xticks(size=18)
plt.yticks(size=18)
plt.show()

```
