### Hard CCS values

```python
def bolt_ccs(ccs, energy):
    """Calculates the Boltzmann weighted average CCS based on given energies.
    """
    g = energy * 627.503
    mn = g.min()
    relG = g - mn    
    b = np.exp(-relG / 0.5924847535)
    w = (b / b.sum()) * len(b)
    
    ws = DescrStatsW(ccs, weights=w, ddof=0)
    return ws.mean
```

```python
index = []
BW = []
LE = []
SA = []
ET5 = []
ET2 = []
ET1 = []
ETp5 = []
CREST = []
ISiCLE = []
combo = []

for d in range(len(names)):
    index.append(names[d] + f' {adducts[d]}') #f' [M{adducts[d]}]$^{adducts[d][0]}$')
    
    dfBW = pd.read_csv(f'../data/MC-ccs-results/molid{molids[d]}_MC_ccs_iter1000.csv')
    lastBW = dfBW.loc[dfBW.index[-1]]    
    #dfET = pd.read_csv(f'../data/MC-ccs-results/molid{molids[d]}_MC_BET_iter1000.csv')
    dfET = pd.read_csv(f'../data/MC-ccs-results/molid{molids[d]}_MC_ccs_minthreshold_iter1000.csv')
    lastET = dfET.loc[dfET.index[-1]]
    
    BW.append(lastBW['Boltzmann Weighted'])
    LE.append(lastBW['Lowest Energy CCS'])
    SA.append(lastBW['Simple Ave'])
    
    ET5.append(lastET['Ave Low Threshold 5kcal'])
    ET2.append(lastET['Ave Low Threshold 2kcal'])
    ET1.append(lastET['Ave Low Threshold 1kcal'])
    ETp5.append(lastET['Ave Low Threshold 0.5kcal'])
 

    # Grab CREST (Non-DFT optimized structure, just the CREST best)
    fileCREST = f'../data/crest/tsvs-crest-structure-simple/{key[d]}_{adducts[d]}_0002_geom01.tsv'
    if exists(fileCREST):
        dfCREST = pd.read_csv(fileCREST, delim_whitespace=True)
        CREST.append(dfCREST['ccs'].values[0])
    else:
        CREST.append(np.nan)
        
    # Grab ISiCLE style basline
    fileISiCLE2 = f'../data/dgo_30/newbench0.1.5/{key[d]}_{adducts2[d]}.csv'
    if exists(fileISiCLE2):
        dfISiCLE = pd.read_csv(fileISiCLE2)
    else:
        dfISiCLE = pd.read_csv(f'../data/dgo_30/multconformer_ccs/{key[d]}_{adducts[d]}.tsv', delim_whitespace=True)
    ISiCLE.append(bolt_ccs(dfISiCLE['ccs'].values,dfISiCLE['dft_energy'].values))

    # Grab the best methods-combo CCS
    with open(f'../data/MC-ccs-results/molid{molids[d]}_bestcombo_ccs.txt', 'r') as f:
        z = f.read().splitlines()
        combo.append(float(z[0])) #f.read()) #.splitlines())
        
dfTable = pd.DataFrame([BW, LE, SA, ET5, ET2, ET1, ETp5, CREST, ISiCLE, combo, measured_ccs],
                        index=['BW', 'LE', 'SA',
                               'ET 5', 'ET 2', 'ET 1', 'ET 0.5',
                               'CREST', 'ISiCLE', 'Best Combo',
                               'Experimental CCS'], #, 'SDS 300 (BW)', '30 DFT optimized (BW)']).T    
                        columns=index).T

dfTable.round(1)
```
