```python
# Finds what cycle the ith lowest energy conformer was generated

mols = []
for ID in molids:
    mols.append(f'../data/molid{ID}_ccs.csv')
mols.sort()

molecule = []

for mol in mols:
    molecule.append(mol.split('/')[-1].split('_')[0])

returndf = pd.DataFrame(molecule, columns=['Molecule'])


LEs = [x for x in range(0,10)]

for lowest_n in LEs:
    cycles = []    
    for mol in mols:
        df = pd.read_csv(mol)
        df = df.sort_values(by='dft_energy')

        # Grab the cycle of the lowest energy conformer                                                                                        
        cycles.append(df['cycle'].iloc[lowest_n])

    returndf[f'LE {lowest_n}'] = cycles
returndf
```
